# 딥러닝 기본

* 단순 선형 분리의 대표적인 알고리즘이 로지스틱회귀입니다.  
* 로지스틱회귀 모델은 회귀와 분류의 조합으로 이루어져 있습니다. 
* 이런 단순한 선형분리로는 실제 문제에서 해결하지 못하는 경우가 많았고,
* 이를 여러 개의 로지스틱회귀로 구성된 형태로는 문제를 해결할 수 있다는 것을 알았습니다. 
* 이때, 하나의 로지스틱회귀는 독립된 입력과 출력이 있고, 다시 이 출력은 다음 로지스틱회귀의 입력으로 사용되며, 최종출력을 만들어 냅니다. 
* 이것은 인간의 신경세포 뉴런의 동작원리와 유사합니다. 

# 신경망 구조
* 신경 세포 뉴런(neuron)은 이전 뉴런으로부터 입력신호를 받아 또 다른 신호를 발생시킵니다. 
* 그러나 입력에 비례해서 출력을 내는 형태($y=WX$)가 아니라, **입력 값들의 모든 합** 이 어느 **임계점(threshold)** 에 도달해야만 출력 신호를 발생시킵니다. 


* 이처럼 입력신호를 받아 특정 값의 임계점을 넘어서는 경우에, 출력을 생성해주는 함수를 **활성화 함수(Activation Function)** 이라고 합니다. 
* Logistic Regression 시스템의 sigmoid함수가 대표적인 **활성화 함수** 입니다. 



* 뉴런과 비슷한 메커니즘을 사용하면서 인공적으로 '생각'하는 그 무언가를 만들 수 있지 않을까?라는 상상으로 출발.
* 인공 신경망(Artificial Neural Network:ANN) 연구입니다. 
* 신경망을 이루는 가장 중요한 기본 단위는 **퍼셉트론(Perceptron)** 입니다. 
* 퍼셉트론은 입력 값과 활성화 함수를 사용해 출력 값을 다음으로 넘기는 가장 작은 신경망 단위입니다. 

https://playground.tensorflow.org/

## 퍼셉트론(Perceptron)

* $x$는 입력값을 의미하며, $w$는 가중치(Weight), $z$는 출력값입니다. 
* 그림 안의 원은 인공 뉴런에 해당됩니다. 각각의 인공 뉴런에서 보내진 입력값 $x$는 각각의 가중치$w$와 함께 종착지인 인공 뉴런에 전달되고 있습니다.
* 각각의 입력값에는 각각의 가중치가 존재하는데, 이때 가중치의 값이 크면 클수록 해당 입력 값이 중요하다는 것을 의미합니다.

# 신경망 크기

* 인공 신경망에 층을 더해서 심층 신경망으로 만들어 봅시다.
* ANN: 입력층 + 은닉층(1개 이상) + 출력층
* 심층 신경망 또는 딥러닝 (DNN)은 은닉층이 2개 이상인 신경망을 말합니다.

## 텐서의 실제 사례

* 벡터 데이터: (samples, features) 크기의 2D 텐서
* 시계열 데이터 또는 시퀀스sequence 데이터: (samples, timesteps, features) 크기의 3D 텐서
* 이미지: (samples, height, width, channels) 또는 (samples, channels, height, width) 크기의 4D 텐서
* 동영상: (samples, frames, height, width, channels) 또는 (samples, frames, channels, height, width) 크기의 5D 텐서
