```python
# 여러 데이터를 하나로 통합하기 위해 - merge(옆으로) - sql의 join
# 국적코드표가 있는 엑셀파일 불러옴

code_master = pd.read_excel('sample_codemaster.xlsx')
code_master
```

```python
# 왼쪽 테이블 기준으로 결합
sample_1_code = pd.merge(left = sample_1, right = code_master,
how = 'left', left_on = '국적코드',right_on = '국적코드')

sample_1_code
```

```python
# how = inner => 두 테이블의 기준 칼럼 값이 서로 일치하는 경우에만 데이터를 통합
sample_1_code = pd.merge(left = sample_1, right = code_master,
how = 'inner', left_on = '국적코드',right_on = '국적코드')

sample_1_code
```

```python
# 여러 데이터를 하나로 통합하기 위해 - append(아래로)
sample_2 = pd.read_excel('sample_2.xlsx', header=1, skipfooter=2, usecols='A:C') # 데이터불러오기
sample_2.head(3)
```

```python
sample_2['기준년월'] = '2020-12'

sample_2_code = pd.merge(left = sample_2, right = code_master, how = 'left', left_on = '국적코드',right_on = '국적코드')

sample_2_code
```

```python
# sample_2_code 작성 후 두 데이터를 아래로 통합
sample = sample_1_code.append(sample_2_code, ignore_index = True)
sample
```

```python
sample.to_excel('sample230214.xlsx', index=False)
```

```python
# 데이터 집계 - 피벗테이블: 기존 데이터의 칼럼을 재구성해서 데이터에 대한 통계를 한눈에 파악할 수 있게 정리한 표
sample_pivot = sample.pivot_table(values = '입국객수', #sample 데이터를 피벗화
                                  index = '국적명',  # 피벗테이블에 들어갈 값
                                  columns='기준년월', # 피벗 칼럼에 기준년월 배치
                                  aggfunc = 'mean') # 입국객수를 국정명과 기준년월 기준으로 평균을 계산
```

```python
sample_pivot
```

```python
# index, colums 는 항상 있어야 하는 것은 아님
sample_pivot = sample.pivot_table(values = '입국객수', 
                                  index = '국적명', 
                                  columns='기준년월', 
                                  aggfunc = 'max') # 국적에 따른 최대 입국자수
```
